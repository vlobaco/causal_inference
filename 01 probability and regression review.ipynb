# Probability and Regression Review

## Regression Line

```python
import numpy as np
from matplotlib import pyplot as plt

# Number of observations
n_observations = 10000

# Generate random data for x and u from a normal distribution
x = np.random.normal(0, 1, size=n_observations)
u = np.random.normal(0, 1, n_observations)

# Generate y using a linear relationship with x and some noise u
y = 5.5 * x + 12 * u

# Calculate the means of x and y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate the sample covariance between x and y
sample_cov = np.mean((x - x_mean) * (y - y_mean))

# Calculate the sample variance of x
sample_var = np.mean((x - x_mean) ** 2)

# Estimate the slope (beta1_hat) of the regression line
beta1_hat = sample_cov / sample_var

# Estimate the intercept (beta0_hat) of the regression line
beta0_hat = y_mean - beta1_hat * x_mean

# Calculate the predicted values of y using the regression line
y_hat = beta0_hat + beta1_hat * x

# Plot the original data points
plt.scatter(x, y)

# Plot the regression line
plt.plot(x, y_hat, color='red')

# Add labels and title to the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('OLS Regression Line')

# Add a dotted line at x=0 and y=0
plt.axvline(x=0, color='black', linestyle='dotted')
plt.axhline(y=0, color='black', linestyle='dotted')

# Add text to display the estimated coefficients
plt.text(-3.75, 25, f'y-intercept = {beta0_hat:.2f}', color='red', weight='bold')
plt.text(1.75, -10, f'Slope = {beta1_hat:.2f}', color='red', weight='bold')

# Display the plot
plt.show()
```

```python
u_hat = y - y_hat

# Plot the residual data points
plt.scatter(x, u_hat)

# Add labels and title to the plot
plt.xlabel('Fitted values')
plt.ylabel('Residuals')

# Add a dotted line at y=0
plt.axhline(y=0, color='black')

# Display the plot
plt.show()
```

```python
# Calculate the residual sum of squares (RSS)
SR = np.sum(u_hat)
SSR = np.sum(u_hat ** 2)
SST = np.sum((y - y_mean) ** 2)
SSE = np.sum((y_hat - y_mean) ** 2)
R2 = 1 - (SSR / SST)

print(f"Residual Sum (SR): {SR:.2f}")
print(f"Residual Sum of Squares (SSR): {SSR:.2f}")
print(f"Total Sum of Squares (SST): {SST:.2f}")
print(f"Explained Sum of Squares (SSE): {SSE:.2f}")
print(f"R-squared (R2): {R2:.2f}")
```
